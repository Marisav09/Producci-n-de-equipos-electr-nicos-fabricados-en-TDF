```python
# carga tabla desde la url https://estudiodelamo.com/cotizacion-historica-dolar-peso-argentina/
import pandas as pd

url = 'https://estudiodelamo.com/cotizacion-historica-dolar-peso-argentina/'
df = pd.read_html(url)

# La función read_html devuelve una lista de DataFrames encontrados en la página.
# Es posible que la tabla que buscas no sea la primera.
# Puedes inspeccionar la lista 'df' para encontrar la tabla correcta.
# Por ejemplo, para mostrar la primera tabla:
if df:
  print(df[0])
else:
  print("No se encontraron tablas en la URL.")
```

```python
# pasar a excel
if df:
  df[0].to_excel('cotizacion_dolar.xlsx', index=False)
  print("La tabla se ha guardado como 'cotizacion_dolar.xlsx'")
else:
  print("No hay tabla para guardar.")
```

```python
# muestra dataframe df solo las 6 filas
df[0].head(5)
```

```python
# crear nuevo dataframe eliminar columna Variación anual dólar oficial
df2 = df[0].drop(columns=['Variación anual dólar oficial (según último mes del año)'])
df2.head(5)
```

```python
# eliminar las 2 primeras filas
df2 = df2.iloc[2:]
df2.head()
```

```python
print("Columnas de df2:", df2.columns)
# cambiar nombre de columna "Unnamed: 0" a "Año
df2 = df2.rename(columns={'Unnamed: 0': 'Año'})
print("Columnas de df2 después del cambio:", df2.columns)
```

```python
# pasar las columnas a filas
# 2. Aplicar pd.melt() para transformar el DataFrame
# id_vars: 'Año' es la columna que queremos mantener fija como identificador.
# var_name: El nombre que queremos darle a la nueva columna que contendrá los nombres de los meses.
# value_name: El nombre que queremos darle a la nueva columna que contendrá los valores.
df_transformado = pd.melt(df2,
                          id_vars=['Año'],
                          var_name='Mes',
                          value_name='Cotizacion') # Usé 'Cotizacion' basado en tu segunda imagen

# 3. Ordenar el DataFrame para que los meses estén en el orden correcto
# Primero, creamos una lista del orden deseado de los meses.
orden_meses = ['ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']

# Convertimos la columna 'Mes' a un tipo de dato categórico con el orden especificado.
df_transformado['Mes'] = pd.Categorical(df_transformado['Mes'], categories=orden_meses, ordered=True)

# Ordenamos el DataFrame por 'Año' y luego por 'Mes'.
df_final = df_transformado.sort_values(by=['Año', 'Mes']).reset_index(drop=True)
# eliminar filas con vlaores nulos
df_final = df_final.dropna()

print("DataFrame Transformado (como la segunda imagen, o muy similar):")
print(df_final)
```

```python
# pasar a excel
# Usamos .empty para verificar si el DataFrame no está vacío antes de intentar guardarlo.
if not df_final.empty:
  df_final.to_excel('cotizacion_dolar-Preprocesado.xlsx', index=False)
  print("La tabla se ha guardado como 'cotizacion_dolar-Preprocesado.xlsx'")
else:
  print("No hay tabla para guardar.")
```
