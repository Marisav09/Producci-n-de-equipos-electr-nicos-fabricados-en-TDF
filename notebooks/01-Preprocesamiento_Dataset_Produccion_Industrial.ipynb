## Extracción del Dataset principal.

Este dataset muestra la producción de equipos electrónicos fabricados en TDF

```python
import pandas as pd

file_name = "/content/drive/MyDrive/Colab Notebooks/Proyecto AprendAutom/14_3_03_Produccion_Industrial-1-1.xlsx"

df_produccion = pd.read_excel(file_name, sheet_name="ELECTRONICA", header=3)
df_produccion = df_produccion.rename(columns={
        'Unnamed: 0': 'Año',
        'Unnamed: 1': 'Mes',
        'Unnamed: 2': 'Total'})
df_produccion = df_produccion.iloc[2:].copy()
df_produccion = df_produccion.iloc[:-5].copy()
df_produccion
```

```python
df_produccion.columns = df_produccion.columns.str.strip()

# Columnas a conservar
columnas_deseadas = [
    'Año',
    'Mes',
    'Total',
    'Acondicionador de Aire',
    'Equipo de Radiocomunicaciones Móviles Celulares',
    'Horno/Cocina a Microondas',
    'Televisor',
    'Módulo Electrónico para Automotor']
df_produc_features = df_produccion[columnas_deseadas]
# Renombrar las columnas
df_produc_features = df_produc_features.rename(columns={
        'Acondicionador de Aire': 'Aireacondicionado',
        'Equipo de Radiocomunicaciones Móviles Celulares': 'Celulares',
        'Horno/Cocina a Microondas': 'Microondas',
        'Módulo Electrónico para Automotor': 'Mod Electr Autom'})
print("DataFrame 'df_produc_features' con solo las columnas deseadas:")
print(df_produc_features)
```

```python
# rellenar los valores del año correspondiente
df_produc_features['Año'] = df_produc_features['Año'].ffill()
df_produc_features['Año'] = df_produc_features['Año'].astype(int)

# Cambiar el tipo de datos de las columnas restantes
columnas_a_entero = [
    'Total',
    'Aireacondicionado',
    'Celulares',
    'Microondas',
    'Televisor',
    'Mod Electr Autom' ]

for col in columnas_a_entero:
    if col in df_produc_features.columns:
        df_produc_features[col] = df_produc_features[col].astype(int)
    else:
        print(f"Advertencia: La columna '{col}' no se encontró en el DataFrame.")

print("\nDataFrame 'df_produc_features' después del cambio de tipos:")
print(df_produc_features)
```

```python
print("\nInformación del DataFrame 'df_produc_features':")
print(df_produc_features.info())
```

```python
# Guardar dataset generado
if not df_produc_features.empty:
  df_produc_features.to_excel('Produccion_Industrial-Preprocesado.xlsx', index=False)
  print("La tabla se ha guardado como 'Produccion_Industrial-Preprocesado.xlsx'")
else:
  print("No hay tabla para guardar.")
```
