```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
import warnings

warnings.filterwarnings('ignore') # Para ignorar advertencias de fechas futuras con pd.to_datetime

# df_final['Fecha'] = pd.to_datetime(df_final['Año'].astype(str) + '-' + df_final['Mes'].astype(str) + '-01')
# df_final = df_final.set_index('Fecha')
# df_final = df_final.sort_index()



# --- PASO 1: Crear un índice de tiempo ---
df_final['Fecha'] = pd.to_datetime(df_final['AÑO'].astype(str) + '-' + df_final['Num_Mes'].astype(str) + '-01')
df_final = df_final.set_index('Fecha')
df_final = df_final.sort_index()

print("--- df_final con DatetimeIndex y ordenado (¡sin NaN en datos históricos!) ---")
print(df_final.info())
print(df_final.head())
print(df_final.tail())
print("\n" + "="*80 + "\n")

# --- PASO 2: Remuestrear a frecuencia trimestral ---
# Ahora que sabemos que no hay NaNs en las series mensuales originales (excepto el objetivo),
# el remuestreo de Inflacion y Dolar debería ser limpio.
df_final_trimestral = df_final.resample('QS').agg({
    'Aireacondicionado': 'sum',
    'Indice general': 'mean',
    'Cotizacion': 'mean'
})

print("\n--- df_final_trimestral (limpio en exógenas, NaN en objetivo futuro) ---")
print(df_final_trimestral.info())
print(df_final_trimestral.head())
print(df_final_trimestral.tail())
print(f"NaNs en Aireacondicionado (después de remuestreo): {df_final_trimestral['Aireacondicionado'].isna().sum()}")
print(f"NaNs en Inflacion_Indice (después de remuestreo): {df_final_trimestral['Indice general'].isna().sum()}")
print(f"NaNs en Cotizacion_Dolar (después de remuestreo): {df_final_trimestral['Cotizacion'].isna().sum()}")
print("\n" + "="*80 + "\n")

# --- PASO 3: Manejo de NaN (solo para la variable objetivo 'Aireacondicionado' en el futuro) ---
# Las otras columnas ya sabemos que no tienen NaNs.
# La columna 'Aireacondicionado' tendrá NaNs solo en los períodos que queremos predecir.
# No necesitamos rellenar NaN en el historial aquí, ya que asumimos que no hay.

# --- PASO 4: Ingeniería de Características para sklearn ---
TARGET = 'Aireacondicionado'
df_features = df_final_trimestral.copy()

# 1. Característica de Tiempo (Tendencia)
df_features['trimestre_num'] = np.arange(len(df_features))

# 2. Característica de Estacionalidad (Trimestre del año)
df_features['trimestre_del_año'] = df_features.index.quarter
df_features = pd.get_dummies(df_features, columns=['trimestre_del_año'], prefix='Q')

# 3. Lags de la Variable Objetivo (Producción de Aire Acondicionado)
# Estos lags deben ser creados sobre la serie original (que tiene NaNs al final)
# para que los futuros NaN en los lags se manejen en la predicción recursiva.
df_features['lag_1_ac'] = df_features[TARGET].shift(1)
df_features['lag_4_ac'] = df_features[TARGET].shift(4)

# 4. Lags de Variables Exógenas
# Como Inflacion_Indice y Cotizacion_Dolar no tienen NaNs, sus lags también estarán limpios.
df_features['lag_1_inflacion'] = df_features['Indice general'].shift(1)
df_features['lag_1_dolar'] = df_features['Cotizacion'].shift(1)

# Eliminar las primeras filas que ahora tienen NaN debido a los lags (ej. los primeros 4 trimestres)
# Esto solo aplica al conjunto de entrenamiento. Para la predicción, los NaNs en los lags
# para el futuro serán rellenados por el valor predicho recursivamente.
df_features_clean = df_features.dropna(subset=df_features.columns.drop(TARGET)) # Deja NaNs en TARGET si son futuros

print("\n--- df_features con características de ML (listo para dividir) ---")
print(df_features_clean.head())
print(df_features_clean.tail())
print(df_features_clean.info())
print(f"NaNs en Aireacondicionado después de crear features: {df_features_clean['Aireacondicionado'].isna().sum()}")
print("\n" + "="*80 + "\n")

# --- PASO 5: Dividir en Conjuntos de Entrenamiento y Predicción Futura ---

# El último índice con datos reales de Aireacondicionado.
# En la simulación, asumo que los datos van de 2015 a 2021 (7 años).
# Si tu df_final real es diferente, ajusta esta lógica.
last_valid_idx_prod = df_features_clean[TARGET].last_valid_index()

if last_valid_idx_prod is None:
    print("La columna 'Aireacondicionado' no contiene ningún valor no nulo en el período histórico.")
else:
    # Conjunto de entrenamiento: Todas las filas hasta el último índice válido de producción
    train_df = df_features_clean.loc[:last_valid_idx_prod].copy()

    X_train = train_df.drop(columns=[TARGET])
    y_train = train_df[TARGET]

    print(f"Dimensiones de X_train: {X_train.shape}")
    print(f"Dimensiones de y_train: {y_train.shape}")
    print(f"Última fecha de entrenamiento: {X_train.index.max().strftime('%Y-%m-%d')}")

    # Conjunto de predicción futura: Filas después del último índice válido de producción
    # Estas son las filas donde el TARGET es NaN
    future_predict_df = df_features.loc[df_features.index > last_valid_idx_prod].copy()

    print(f"Dimensiones de future_predict_df (antes de la predicción recursiva): {future_predict_df.shape}")
    print(f"Primer fecha a predecir: {future_predict_df.index.min().strftime('%Y-%m-%d') if not future_predict_df.empty else 'N/A'}")
    print(f"Última fecha a predecir: {future_predict_df.index.max().strftime('%Y-%m-%d') if not future_predict_df.empty else 'N/A'}")

    # --- PASO 6: Entrenar el Modelo de sklearn ---
    print("\n--- Entrenando RandomForestRegressor ---")
    model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)
    print("Modelo entrenado exitosamente.")

    # --- PASO 7: Realizar Predicciones Futuras (Recursiva) ---
    if not future_predict_df.empty:
        print("\n--- Realizando predicciones recursivas ---")

        # DataFrame para guardar todas las predicciones (incluyendo los valores reales para los lags)
        # Inicialmente, llenamos con los valores reales conocidos
        all_predictions_series = df_features_clean[TARGET].copy()

        for i, current_quarter_index in enumerate(future_predict_df.index):
            # Obtener las características para el trimestre actual a predecir
            # Necesitamos una copia para evitar SettingWithCopyWarning
            features_for_pred = df_features.loc[[current_quarter_index]].copy()

            # Rellenar los lags de 'Aireacondicionado' con los valores previos (reales o predichos)
            # lag_1_ac: valor del trimestre anterior
            prev_quarter_index = current_quarter_index - pd.DateOffset(months=3)
            if prev_quarter_index in all_predictions_series.index:
                features_for_pred['lag_1_ac'] = all_predictions_series.loc[prev_quarter_index]
            else:
                # Si el trimestre anterior está fuera del rango de all_predictions_series, rellena con 0 o la media del entrenamiento
                features_for_pred['lag_1_ac'] = X_train['lag_1_ac'].mean() # O 0, dependiendo del contexto

            # lag_4_ac: valor del mismo trimestre del año anterior
            prev_year_quarter_index = current_quarter_index - pd.DateOffset(years=1)
            if prev_year_quarter_index in all_predictions_series.index:
                features_for_pred['lag_4_ac'] = all_predictions_series.loc[prev_year_quarter_index]
            else:
                 features_for_pred['lag_4_ac'] = X_train['lag_4_ac'].mean() # O 0

            # Asegurarse de que no queden NaNs en las características (ej. al inicio del future_predict_df)
            # Esto es importante si el primer trimestre a predecir no tiene lags disponibles de X_train.
            features_for_pred = features_for_pred.fillna(method='ffill').fillna(method='bfill').fillna(0) # Ffill/Bfill dentro de la fila, luego 0 si aún hay NaNs

            # Asegurar que el orden de las columnas sea el mismo que el de entrenamiento
            features_for_pred = features_for_pred[X_train.columns]

            # Hacer la predicción
            predicted_value = model.predict(features_for_pred)[0]

            # Asegurar que la producción no sea negativa
            if predicted_value < 0:
                predicted_value = 0

            # Guardar la predicción para ser usada en los lags de futuras predicciones
            all_predictions_series.loc[current_quarter_index] = predicted_value

        # Filtrar solo las predicciones para el período futuro
        future_predictions = all_predictions_series.loc[future_predict_df.index]

        print("\n--- Predicciones de Producción de Aire Acondicionado para el Futuro ---")
        print(future_predictions)

        # --- Visualizar Resultados ---
        plt.figure(figsize=(18, 8))
        plt.plot(y_train.index, y_train, label='Datos de Entrenamiento (Reales)', color='blue')
        plt.plot(future_predictions.index, future_predictions, label='Predicción Futura', color='red', linestyle='--')

        plt.title('Producción Trimestral de Aire Acondicionado con Predicción ML (Random Forest)')
        plt.xlabel('Fecha')
        plt.ylabel('Producción')
        plt.legend()
        plt.grid(True)
        plt.show()

    else:
        print("No hay trimestres futuros para predecir.")
```
