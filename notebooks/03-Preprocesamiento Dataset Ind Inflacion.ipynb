## Extracción del Dataset Inflacion.
Consiste en 3 datasets, los cuales muestran la serie historica del indice de Inflacion (IPC) en diferentes períodos.

En este etapa se unirán los 3 datasets para formar un único dataset que muestre esta información.

```python
import pandas as pd
file_name_ipc = "/content/drive/MyDrive/Colab Notebooks/Proyecto AprendAutom/sh_ipc_2008.xls"
df_ipc = pd.read_excel(file_name_ipc, sheet_name="Serie Histórica", header=4)
```

```python
columnas_a_mantener = ['año', 'mes', 'Nivel general']
df_ipc_2013 = df_ipc[columnas_a_mantener]
df_ipc_2013 = df_ipc_2013.iloc[840:852].copy()
df_ipc_2013 = df_ipc_2013.reset_index(drop=True)
df_ipc_2013['mes'] = df_ipc_2013['mes'].astype(int)
df_ipc_2013['año'] = df_ipc_2013['año'].astype(int)
df_ipc_2013['Nivel general'] = df_ipc_2013['Nivel general'].astype(float)

print("DataFrame 'df_ipc_filtrado' después de seleccionar columnas y filtrar filas:\n")
print(df_ipc_2013)
print("\nColumnas del DataFrame filtrado:")
print(df_ipc_2013.dtypes)
```

```python
file_name_ipc2 = "/content/drive/MyDrive/Colab Notebooks/Proyecto AprendAutom/sh_ipcnu.xls"
df_ipc2 = pd.read_excel(file_name_ipc2, sheet_name="Serie Histórica", header=4)
df_ipc2
```

```python
columnas_a_mantener2 = ['Año', 'Mes', 'Nivel general']
df_ipc_2014_2016 = df_ipc2[columnas_a_mantener2]
df_ipc_2014_2016 = df_ipc_2014_2016.iloc[1:23].copy()
df_ipc_2014_2016 = df_ipc_2014_2016.reset_index(drop=True)
df_ipc_2014_2016.iloc[17, 0] = 2015
df_ipc_2014_2016.iloc[18, 0] = 2015
df_ipc_2014_2016
```

```python
df_ipc_2014_2016['Mes'] = df_ipc_2014_2016['Mes'].astype(int)
df_ipc_2014_2016['Año'] = df_ipc_2014_2016['Año'].astype(int)
df_ipc_2014_2016['Nivel general'] = df_ipc_2014_2016['Nivel general'].astype(float)

df_ipc_2014_2016.dtypes

print("DataFrame después de seleccionar columnas y filtrar filas:\n", df_ipc_2014_2016.head())
print("\nColumnas del DataFrame filtrado:\n", df_ipc_2014_2016.dtypes)
```

```python
file_name_ipc2 = "/content/drive/MyDrive/Colab Notebooks/Proyecto AprendAutom/sh_ipc_05_25.xls"
df_ipc3 = pd.read_excel(file_name_ipc2, sheet_name="Indices IPC Cobertura Nacional", header=5)
df_ipc3.head()
```

```python
df_ipc3 = df_ipc3.iloc[3:4].copy()
df_ipc3
```

```python
df_transposed = df_ipc3.set_index('Total nacional').T
df_final_ipc_nacional = df_transposed.reset_index()
# Renombrar las columnas para mayor claridad
df_final_ipc_nacional.columns = ['Fecha', 'Nivel general']
#print(df_final_ipc_nacional.head())
df_final_ipc_nacional['Año'] = df_final_ipc_nacional['Fecha'].dt.year
df_final_ipc_nacional['Mes'] = df_final_ipc_nacional['Fecha'].dt.month
columnas_a_mantener = ['Año', 'Mes', 'Nivel general']
df_ipc_2017_2025=df_final_ipc_nacional[columnas_a_mantener]
df_ipc_2017_2025.head()

```

```python
df_ipc_2017_2025['Mes'] = df_ipc_2017_2025['Mes'].astype(int)
df_ipc_2017_2025['Año'] = df_ipc_2017_2025['Año'].astype(int)
df_ipc_2017_2025['Nivel general'] = df_ipc_2017_2025['Nivel general'].astype(float)
df_ipc_2017_2025.dtypes
```

```python
print(df_ipc_2013.head())
print(df_ipc_2014_2016.head())
print(df_ipc_2017_2025.head())
#
```

## Unificación de los datates

```python
df_ipc_2013 = df_ipc_2013.rename(columns={'año': 'Año'})
df_ipc_2013 = df_ipc_2013.rename(columns={'mes': 'Mes'})
df_ipc_unificado = pd.concat([df_ipc_2013, df_ipc_2014_2016, df_ipc_2017_2025], ignore_index=True)
df_ipc_unificado.head()
```

```python
print("DataFrame UNIFICADO (df_ipc_unificado) - Primeras 10 filas:")
print(df_ipc_unificado.head(10))
print("\nDataFrame UNIFICADO (df_ipc_unificado) - Últimas 10 filas:")
print(df_ipc_unificado.tail(10))
print("\nInformación general del DataFrame UNIFICADO:")
print(df_ipc_unificado.info())
print("\nColumnas y tipos de datos del DataFrame UNIFICADO:")
print(df_ipc_unificado.dtypes)
```

```python
#Guardar dataset generado
if not df_ipc_unificado.empty:
  df_ipc_unificado.to_excel('Indice de Inflacion (ipcn)-Preprocesado.xlsx', index=False)
  print("La tabla se ha guardado como 'Inflacion (ipcn)-Preprocesado.xlsx'")
else:
  print("No hay tabla para guardar.")
```
