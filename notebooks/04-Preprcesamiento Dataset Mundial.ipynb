## Extracción del Dataset Mundial.
Este dataset cobrará relevancia al momento de modelar la fabricación de televisores.  

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

url = 'https://www.goal.com/es-ar/noticias/todas-las-finales-de-los-mundiales/8k21jcs1v7321xkkfde58ectl'

df_list = pd.read_html(url)
df = df_list[0]

# Filtrar filas y columnas
df = df[["Fecha"]]
df = df.loc[19:,:]
df
```

```python
# Convertir la columna Fecha a tipo datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')

df['Año'] = df['Fecha'].dt.year
df['Mes'] = df['Fecha'].dt.month
df=df.drop(columns=['Fecha'])
df=df[["Año","Mes"]]

print(df)
```

```python
# Crear un DataFrame con años del 2014 al 2025 y meses del 1 al 12
years = range(2014, 2026)
months = range(1, 13)

# Crear todas las combinaciones de año y mes
df_mundial = pd.MultiIndex.from_product([years, months], names=['Año', 'Mes']).to_frame(index=False)

# Agregar columna "Mundial"
df_mundial['Mundial'] = df_mundial.apply(lambda row: 1 if (row['Año'], row['Mes']) in zip(df['Año'], df['Mes']) else 0, axis=1)

df_mundial.head(10)
```

```python
# Guardar dataset generado
df_mundial.to_excel('mundiales.xlsx', index=False)
```
